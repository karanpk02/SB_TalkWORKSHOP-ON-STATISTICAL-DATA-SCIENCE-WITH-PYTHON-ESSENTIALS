##**Probability and Statistics**
*********
************

The Bernoulli distribution is a discrete probability distribution for a random variable which takes the value 1 with probability

p and the value 0 with probability
1p. It is used to model binary outcomes, such as a coin toss where 1 might represent heads and 0 might represent tails.

####**Example Code: Bernoulli Distribution Simulation**
Below is the example code to simulate a Bernoulli distribution using a **fair coin** (where

=
0.5
p=0.5) and a **biased coin** (where

=
0.3
p=0.3). This code will perform 10000 coin flips for each experiment and repeat the experiment 1000 times to collect results.


````python
import numpy as np
# Fair coin
def experiment(n, p):
    return sum(np.random.choice([0, 1], p=[1 - p, p], size=n))

# Experiment
results = [experiment(10000, 0.3) for i in range(1000)]

# Display results
print("Mean of results:", np.mean(results))
print("Variance of results:", np.var(results))
````

````python
from matplotlib import pyplot as plt
# Create the plot
fig, ax = plt.subplots()
ax.hist(results, bins=20)
# Show the plot
plt.show()
````

````python
import numpy as np

# Define Bernoulli trial function
def bernoulli(p):
    return np.random.choice([0, 1], p=[1 - p, p])

# Define Binomial distribution function
def binomial(n, p):
    return sum([bernoulli(p) for _ in range(n)])

# Generate results from Binomial distribution
results = np.array([binomial(20, 0.5) for _ in range(10000)])

# Display the results
print("Mean of results:", np.mean(results))
print("Variance of results:", np.var(results))

# Plotting the results
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.hist(results, bins=range(22), edgecolor='black', align='left')
ax.set_title('Binomial Distribution Histogram')
ax.set_xlabel('Number of successes')
ax.set_ylabel('Frequency')

plt.show()
````
